```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/99. analyzes/csv/all_hcp_recipients_aggregated_sum.csv')
```

```python
df.head()
```

```python
df['firstname'] = df['name'].apply(lambda x: x.split(' ')[0])
df['lastname'] = df['name'].apply(lambda x: x.split(' ')[len(x.split(' ')) - 1])

df_firstname = df.groupby('firstname').agg({'name': 'count'})
df_firstname = df_firstname.sort_values('name', ascending=False)
df_firstname.head()
```

```python
df_lastname = df.groupby('lastname').agg({'name': 'count'})
df_lastname = df_lastname.sort_values('name', ascending=False)
df_lastname.head()
```
