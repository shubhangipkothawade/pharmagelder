```python
from tabulator import Stream
from glob import glob
    
header = None
aggregate = []
for filename in glob('data/ch/raw_xlsx/*.xlsx'):
    print ("Converting %s" % filename)
    with Stream(filename, headers=1) as stream:
        slug = filename.split('/')[-1].split('.')[0]
        if header is None:
            header = [ 'slug' ]
            header.extend(stream.headers)
            aggregate.append(header)
        withslug = []
        for row in stream.iter():
            if row is not None:
                withslug = [ slug ]
                withslug.extend(row)
                aggregate.append(withslug)
        stream.reset()
        stream.save("data/ch/raw_csv/%s.csv" % slug)
    
with Stream(aggregate) as stream:
    stream.save("data/ch/checkrows.csv")
    print ("Done saving stream.")
```
